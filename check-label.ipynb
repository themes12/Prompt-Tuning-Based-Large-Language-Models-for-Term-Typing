```python
import pandas as pd
from sklearn.model_selection import train_test_split
import json
```

```python
f = open('./datasets/Geonames/label_mapper.json')

data = json.load(f)
```

```python
data
```

```python
label = []
for key in data:
    value = data[key]
    # print("The key and value are ({}) = ({}), type = {}".format(key, value, type(value)))
    if type(value) is dict:
        label.append(value["name"])
    else:
        label += value

print(label)
print(len(label))
```
